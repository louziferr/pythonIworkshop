# Plotting with PyPlot

```python
import matplotlib.pyplot as plt
# make sure to embed the plots inside the notebook
%matplotlib inline    
```

The top-level container for all plot elements are **figures**.

A figure can contain one or more **axes** which set the coordinate system.

There are two ways to use matplotlib:

1) Object oriented style: Expilcitly create figures and axes and call methods on them:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,4,1,8,1])
```

2) Pyplot style: Just use pyplot functions and let pyplot create and manage figures and axes automatically.

```python
plt.plot([1,2,3,4,5], [2,4,1,8,1])
```

```python
x = list(range(0,10))

fig, ax = plt.subplots()

ax.plot(x, x, label='linear')
ax.plot(x, [xi**2 for xi in x], marker="x", linestyle="--", label='quadratic')
ax.plot(x, [xi**3 for xi in x], label='cubic')
ax.scatter([1,2,3],[300,200,500])

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_xticks(x)

ax.set_title("Simple Plot")

ax.legend()
ax.grid()
```

Pyplot style

```python
plt.plot(x, x, label='linear')
plt.plot(x, [xi**2 for xi in x], marker="x", linestyle="--", label='quadratic') 
plt.plot(x, [xi**3 for xi in x], label='cubic')
plt.scatter([1,2,3],[300,200,500])

plt.xlabel('x label')
plt.ylabel('y label')
plt.xticks(x)

plt.title("Simple Plot")

plt.legend()
plt.grid()
```

## Subplots

from https://matplotlib.org/gallery/subplots_axes_and_figures/subplot.html

```python
import numpy as np
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
```

## Bar charts

from 
https://matplotlib.org/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = range(len(labels))  # the label locations
width = 0.35  # the width of the bars
x1 = [xi - width/2 for xi in x]
x2 = [xi + width/2 for xi in x]

fig, ax = plt.subplots()
rects1 = ax.bar(x1, men_means, width, label='Men')
rects2 = ax.bar(x2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

## Scatter plots

```python
cloud_1 = np.random.multivariate_normal([3,2], [[2,0],[0,3]], size=50)
cloud_2 = np.random.multivariate_normal([5,0], [[0.5,0],[0,0.4]], size=50)

x1 = cloud_1[:,0]
y1 = cloud_1[:,1]

x2 = cloud_2[:,0]
y2 = cloud_2[:,1]
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.scatter(x1, y1, marker="o", s=50, color="red",alpha=0.8, label="Cloud 1")
ax.scatter(x2, y2, marker="x", s=50, color="green",alpha=0.8, label="Cloud 2")

ax.set_xlabel("x", fontsize=13)
ax.set_ylabel("y", fontsize=13)

plt.title("Clouds")

plt.grid()
ax.legend()
```

## Histograms

How to construct a histogram: 

1) Devide entire the range of values into series of intervals ( =bins ) and count how many values lie in each interval. 

2) Plot the values on the horizontal axis and the count on the vertical axis. 

```python
data = np.random.normal(5.0, 2.0, size=1000)

# plot the histogram, choose the number of bins
count, bins, patches = plt.hist(data, bins=10)
```
